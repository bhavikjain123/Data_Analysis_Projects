```python
import pandas as pd

data = {'variable1': [1, 2, 3, 4, 5],
        'variable2': [2, 3, 4, 5, 6]}

df = pd.DataFrame(data)

# Calculate the Pearson correlation coefficient
correlation_matrix = df.corr()
pearson_coefficie = correlation_matrix.loc['variable1', 'variable2']

print("Pearson correlation coefficient:", pearson_coefficie)
```

```python
#correlation betn variables 
corr_df = df_tracks.drop(["key", "explicit", "mode"], axis=1).corr(method = "pearson")
plt.figure(figsize=(14,6))
heatmap = sns.heatmap(corr_df, annot = True, vmin= -1, vmax=1, cmap="inferno",  fmt = ".1g", linewidths = 3, linecolor= "Black")
heatmap.set_title("Correlation Heatmap Between Variable")
```

```python
df_tracks['Dates']= df_tracks.index.get_level_values('release_date')
df_tracks.Dates = pd.to_datetime(df_tracks.Dates)
Years = df_tracks.Dates.dt.year

#months = df_tracks.Dates.dt.month

print(Years)
```

```python
sns.displot( Years, kind = "hist", discrete= True, aspect=2,height=5).set(title="Number of songs per year")
```

```python
total_dr= df_tracks.duration
plt.figure(figsize=(18,7))
sns.barplot(x=Years, y= total_dr, errwidth = False).set(title="Years vs Duration")
plt.xticks(rotation = 90)
```

```python
total_dr= df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=Years, y= total_dr,ax = ax, errwidth = False).set(title="Years vs Duration")
plt.xticks(rotation = 90)
```

```python
#understanding error bar
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4]
y = [10, 15, 7, 12]
errors = [1, 2, 0.5, 1]

# Create a basic plot with error bars
plt.errorbar(x, y, yerr=errors, fmt='o', capsize=5)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plot with Error Bars')

# Show the plot
plt.show()
```

```python
total_dr = df_tracks.duration
sns.set_style(style = "whitegrid")
plt.figure(figsize=(10,6))
sns.lineplot(x= Years, y= total_dr).set(title="Year vs duration")
plt.xticks(rotation=60)
```

```python
sns.barplot(x="duration_ms", y="genre", data = df_genre, errwidth = False).set(title="durations of songs in different genres")
plt.xlabel("Duration in mili-seconds")
plt.ylabel("Genres")
```

```python
sns.barplot(x="duration_ms", y="genre", data = df_genre).set(title="durations of songs in different genres")
sns.color_palette("rocket", as_cmap= True)
plt.xlabel("Duration in mili-seconds")
plt.ylabel("Genres")
```

```python
sns.set_style(style= "darkgrid")
plt.figure(figsize=(10,5))
famous = df_genre.sort_values("popularity", ascending = False).head()
sns.barplot(x = "popularity", y="genre", data = famous).set(title="Top genres by popularity")
```
